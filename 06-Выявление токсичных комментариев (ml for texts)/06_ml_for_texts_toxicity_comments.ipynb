### Лемматизация и очистка текстов

```python
def lemmatize_clear(text):
    w = WordNetLemmatizer()
    #m = Mystem()
    text = text.lower() #приведём все слова к нижнему регистру
    word_list = nltk.word_tokenize(text) #разделим на токены
    lemm_text = " ".join([w.lemmatize(word) for word in word_list]) #выделим леммы
    cl_text = re.sub(r'[^a-zA-Z]', ' ', lemm_text) #оставим в текстах только нужные символы
    return " ".join(cl_text.split())

toxic_comments['lemm_text'] = toxic_comments['text'].apply(lemmatize_clear)
```

Похоже, модель чаще ошибается в пользу токсичности комментария. Попробуем перебрать пороги классификации и подобрать тот, при котором f1 будет наиболее высоким. Для удобства создадим функцию, перебирающую пороги и возвращающую предсказания при наибольшем значении метрики.

```python
def def_threshold(model, features, target):
    best_predictions = []
    max_f1 = 0
    best_threshold = 0.5
    for threshold in np.arange(0.01, 1, 0.01):
        probabilities = model.predict_proba(features)
        predictions = (probabilities[:, 1] > threshold)*1

        f1 = f1_score(target, predictions)
        if f1 > max_f1: 
            max_f1 = f1 
            best_threshold = threshold
            best_predictions = predictions
    print('max_f1:', max_f1)
    print('best_threshold:', best_threshold)
    return max_f1, best_threshold, best_predictions 
```

#### С балансировкой классов и перебором порогов классификации

```python
f1_lr_balanced_valid, best_threshold_lr_balanced_valid, best_predictions_lr_balanced_valid  = def_threshold(lr_balanced, features_valid, target_valid)
```

```python
confusion_matrix(target_valid, best_predictions_lr_balanced_valid)
```

f1 лучше, но в несколько раз возросло число ложноотрицательных предсказаний.
