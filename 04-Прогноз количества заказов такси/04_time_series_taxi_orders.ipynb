Используем функцию, при помощи которой сможем подбирать размер шага при вычитании рядов и размер выборки, по которой будет вычисляться скользящее среднее.

```python
def make_features(data, max_lag, rolling_mean_size):

    for lag in range(1, max_lag + 1):
        data['lag_{}'.format(lag)] = data['num_orders'].shift(lag)

    data['rolling_mean'] = data['num_orders'].shift().rolling(rolling_mean_size).mean()
    return data
```

## Обучение

Так как нам необходимо подобрать размер шага и размер выборки, а также гиперпараметры моделей. Выделение тренировочной и тестовой выборок мы будем производить для каждой модели отдельно. 

*Функция для подбора шага и размера выборки*

```python
def select_for_fit(data_to_fit, regressor):
    best_model = None
    best_rmse_test = 48
    best_max_lag = 0
    best_rolling_mean_size = 0
    best_rmse_train = 0
    best_mae_test = 0
    best_mae_train = 0
    for max_lag in range(18, 30, 1):
        for rolling_mean_size in range(10, 30, 1):
            data_to_fit = make_features(data_to_fit, max_lag, rolling_mean_size)
            
            train, test = train_test_split(data_to_fit, shuffle=False, test_size=0.1)
            train = train.dropna()
            
            train_features = train.drop('num_orders',axis = 1)
            train_target = train['num_orders']
    
            test_features = test.drop('num_orders',axis = 1)
            test_target = test['num_orders']
    
            model = regressor
            model.fit(train_features,train_target)
    
            predictions_train = model.predict(train_features)
            predictions_test = model.predict(test_features)
            
            rmse_train = mean_squared_error(train_target,predictions_train)**0.5
            mae_train = mean_absolute_error(train_target,predictions_train)
            
            rmse_test = mean_squared_error(test_target,predictions_test)**0.5
            mae_test = mean_absolute_error(test_target,predictions_test)
            
            if rmse_test <48:
                print("RMSE тестовой выборки: {:.2f}".format(rmse_test), '|', 'max_lag:', max_lag, '|', 'rolling_mean_size:', rolling_mean_size)
            
            if rmse_test < best_rmse_test:
                best_rmse_test = rmse_test
                best_rmse_train =  rmse_train
                best_mae_test = mae_test
                best_mae_train = mae_train
                best_model = model
                best_max_lag = max_lag
                best_rolling_mean_size = rolling_mean_size
    print()
    print("лучшая RMSE обучающей выборки: ", best_rmse_train)
    print("лучшая MAE обучающей выборки: ", best_mae_train)
    print()
    print("лучшая RMSE тестовой выборки: ", best_rmse_test)
    print("лучшая MAE тестовой выборки: ", best_mae_test)
    print()
    print('max_lag:', best_max_lag)
    print('rolling_mean_size:', best_rolling_mean_size)
    return best_model, best_max_lag, best_rolling_mean_size
```

*Функция для разделения выборки*

```python
def train_test(data_to_fit, regressor, best_max_lag, best_rolling_mean_size):  
    data_to_fit = make_features(data_to_fit, best_max_lag, best_rolling_mean_size)
        
    train, test = train_test_split(data_to_fit, shuffle=False, test_size=0.1)
    train = train.dropna()
    
    train_features = train.drop('num_orders',axis = 1)
    train_target = train['num_orders']

    test_features = test.drop('num_orders',axis = 1)
    test_target = test['num_orders']
    
    return train_features, train_target, test_features, test_target
```

*Функция для подсчёта rmse тренировочной выборки*

```python
def predict_train(best_model, train_features, train_target):
    
    predictions_train = best_model.predict(train_features)
        
    rmse_train = mean_squared_error(train_target,predictions_train)**0.5
    mae_train = mean_absolute_error(train_target,predictions_train)
    
    print("RMSE обучающей выборки:", rmse_train)
    print("MAE обучающей выборки:", mae_train)
        
```

*Функция для подсчёта rmse тестовой выборки*

```python
def predict_test(best_model, test_features, test_target):
    
    predictions_test = best_model.predict(test_features)
    
    rmse_test = mean_squared_error(test_target,predictions_test)**0.5
    mae_test = mean_absolute_error(test_target,predictions_test)
        
    print("RMSE тестовой выборки: ", rmse_test)
    print("MAE тестовой выборки: ", mae_test)

    return predictions_test, rmse_test
```

```python
def result(target, predictions, name_of_model):
    
    result_frame = pd.DataFrame(target)
    result_frame[name_of_model] = predictions

    result_frame['2018-08-20':'2018-08-26'].plot(figsize=(15,7), grid=True)
    plt.title('Целевой признак тестовой выборки и готовое предсказание по предпоследней неделе августа')
    plt.show()
    
    result_frame['2018-08-20':'2018-08-22'].plot(figsize=(15,7), grid=True)
    plt.title('Целевой признак тестовой выборки и готовое предсказание по двум дням в августе')
    plt.show()
    
    return result_frame
```
