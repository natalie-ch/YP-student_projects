#### Заполнение пропусков для данных, закодированных методом OHE

Попробуем заполнить пропуски на основе предсказаний модели. 

Выделим срез данных без пропусков в отдельный датасет.

```python
data_t = data[~data['tenure'].isna()]
```

Целевой признак - tenure

```python
target_t = data_t['tenure']
features_t = data_t.drop('tenure', axis=1)
features_train_t, features_valid_t, target_train_t, target_valid_t = train_test_split(
    features_t, target_t, test_size=0.25, random_state=12345)
```

Выделим признаки среза с пропусками для предсказаний модели, на основе которых, заполним пропуски.

```python
data_t_nan = data[data['tenure'].isna()]
features_t_nan = data_t_nan.drop('tenure', axis=1)
```

Масштабируем признаки.

```python
numeric = ['credit_score', 'age', 'balance', 'estimated_salary']
```

```python
scaler = StandardScaler()
scaler.fit(features_train_t[numeric]) 
features_train_t[numeric] = scaler.transform(features_train_t[numeric])
features_valid_t[numeric] = scaler.transform(features_valid_t[numeric])
features_t_nan[numeric] = scaler.transform(features_t_nan[numeric])
pd.options.mode.chained_assignment = None
```

```python
features_train_t.head()
```

```python
print('target_train')
print(target_train_t.value_counts())

print('target_valid')
print(target_valid_t.value_counts())
```

Всего 11 классов. Попытаемся немного снизить дисбаланс классов. Увеличим тренировочную выборку за счет увеличения в два раза самых малочисленных классов - 10 и 0.

```python
def upsample(features, target, repeat):
    features_10 = features[target == 10]
    features_0 = features[target == 0]
    features_other = features[(target != 1)&(target != 0)]
    target_10 = target[target == 10]
    target_0 = target[target == 0]
    target_other = target[(target != 1)&(target != 0)]

    features_upsampled = pd.concat([features_other] + [features_10] * repeat + [features_0] * repeat)
    target_upsampled = pd.concat([target_other] + [target_10] * repeat + [target_0] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled_t, target_upsampled_t = upsample(features_train_t, target_train_t, 2)
```

```python
from catboost import CatBoostClassifier
model = CatBoostClassifier(random_state=12345)
#fit
model.fit(features_upsampled_t, target_upsampled_t)
#predict
predictions_upsampled_t = model.predict(features_upsampled_t)
predictions_train_t = model.predict(features_train_t)
predictions_valid_t = model.predict(features_valid_t)

print()
print('classification report upsampled')
print(classification_report(target_upsampled_t, predictions_upsampled_t))
print()
print('classification report train')
print(classification_report(target_train_t, predictions_train_t))
print()
print('classification report valid')
print(classification_report(target_valid_t, predictions_valid_t))
print()
print('classification report 0')
predictions_0 = pd.Series(target_valid_t*0, index=target_valid_t.index)
print(classification_report(target_valid_t, predictions_0))
print()
print('classification report median')
predictions_median = pd.Series(target_valid_t.median().round(), index=target_valid_t.index)
print(classification_report(target_valid_t, predictions_median))
```

Одолеть мультиклассовый классификатор пока получилось не полностью, но результат лучше, чем у заполнения нулями или медианой.

Подготовим столбец для заполнения пропусков.

## Борьба с дисбалансом

Сначала подготовим данные для увеличения и уменьшения выборки с целью выровнять классы.

### Увеличение выборки

Подготовим данные для обучения моделей. Выравняем классы за счет увеличения выборки. В четыре раза увеличим положительный класс.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones]*repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones]*repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

repeat_upsampling = round((len(features_zeros)/len(features_ones)))
#для OHE
features_upsampled, target_upsampled = upsample(features_train, target_train, repeat_upsampling)
#для Ordinal
features_upsampled_ord, target_upsampled_ord = upsample(features_train_ord, target_train_ord, repeat_upsampling)
```

### Уменьшение выборки

Подготовим данные для обучения моделей. Выравняем классы за счет уменьшения выборки. В четыре раза уменьшим отрицательный класс.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

repeat_downsampling = len(features_ones)/len(features_zeros)
#для OHE
features_downsampled, target_downsampled = downsample(features_train, target_train, repeat_downsampling)
#для Ordinal
features_downsampled_ord, target_downsampled_ord = downsample(features_train_ord, target_train_ord, repeat_downsampling)
```

```python
best_model = None
depth_of_best_model_forest = 0
best_f1_test_forest = 0
for depth in range(1, 10, 1):
    model = RandomForestClassifier(random_state=12345, n_estimators=60, max_depth=depth, class_weight='balanced')
    model.fit(features_plus, target_plus)
    
    predictions_plus = model.predict(features_plus)
    predictions_test = model.predict(features_test)
    
    f1_plus = f1_score(target_plus, predictions_plus)
    f1_test = f1_score(target_test, predictions_test)
    
    if f1_test > best_f1_test_forest:
        best_f1_test = f1_test 
        best_model = model
        depth_of_best_model_forest = depth
        best_f1_plus = f1_plus
        
print("Глубина дерева лучшей модели:", depth_of_best_model_forest)
print("f1_plus лучшей модели:", best_f1_plus)
print("f1_test лучшей модели:", best_f1_test)
```
